```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('cars copy.csv')
df=df[['EngineSize', 'Cylinders', 'Horsepower', 'MPG_City', 'MPG_Highway',
       'Weight', 'Wheelbase', 'Length']]
df.head()
df.describe()
#df.shape[0]=428
```

# plot()
- default is line
- more graphs with kind='line','bar', 'scatter','area','hist'

# 1. Line plot

```python
df.plot()
```

# 2. Scatter plot

```python
df.plot(kind='scatter',
       x='Horsepower', y='Weight')
```

```python
df.plot.scatter(x='EngineSize', y='Weight')
```

# 3. Histogram 

```python
df=df[['EngineSize', 'Cylinders', 'Horsepower', 'MPG_City', 'MPG_Highway', 'Wheelbase', 'Length']]
df.plot(kind='hist')
```

```python
df.plot.hist()
```

# 4. Area plot

```python
#df.plot(kind='area')
df.plot.area()
```

# 5. Bar chart

```python
head=df.head(10)
head.plot.bar()
#df.plot(kind=bar)
```

```python
head.plot.bar(stacked=True)
```

```python
#horizontal bar
head.plot.barh()
```

# 6. Boxplot

```python
box=df.drop('Horsepower', axis=1)
box.plot.box()
```

## hexagonal binning plot

```python
df.plot.hexbin(x='Wheelbase', y='Horsepower',
              gridsize=18)
```

## 7. Density Estimate plot

```python
df.plot.kde()
```

# 8. Pie Plot

```python
df.head(10).plot.pie(y='Cylinders', autopct='%1.1f%%')
```
